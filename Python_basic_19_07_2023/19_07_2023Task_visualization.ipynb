```python
"""
Module contains the Advanced Data Visualization with Pandas and Matplotlib
"""
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno
from sklearn.preprocessing import Normalizer
```

```python
folder_df = pd.read_csv('sample_data/log_csv.csv', sep=",", encoding='Latin-1')
folder_df
```

```python
folder_df.rename(columns = {'Folder Size(KB)':'Folder Size'}, inplace = True)
folder_df.rename(columns = {'File Size(KB)':'File Size'}, inplace = True)
folder_df
```

**Extension wise visulaization**

```python
# Visulaization with seaborn
sns.countplot(x = 'File Extension', data = folder_df)
```

```python
# visualizing the number of files Extension wise visulaization with matplotlib
file_count_df = folder_df.groupby(by=['File Extension'], as_index=False)['File Name'].count()
file_count_df.rename(columns = {'File Name':'File Extension Count'}, inplace = True)
file_count_df
```

```python
plt.bar(x = file_count_df['File Extension'] , height=file_count_df['File Extension Count'],
        color=['black', 'red', 'green', 'blue', 'cyan'], edgecolor="black", width=0.5)
plt.xlabel("Extensions")
plt.ylabel("Extension Count")
plt.show()
```

```python
fig, ax = plt.subplots()
# plt.pie(file_count_df['Unit Count'], labels = file_count_df['Unit'] )
ax.pie(file_count_df['File Extension Count'], labels = file_count_df['File Extension'], autopct='%1.1f%%')
plt.show()
```

**Number OF Files in KB and MB visualization**

```python
# Visulaization with seaborn
sns.countplot(x = 'Unit', data = folder_df)
```

```python
# visualizing the number KB and MB units visulaization with matplotlib
file_count_df = folder_df.groupby(by=['Unit'], as_index=False)['File Name'].count()
file_count_df.rename(columns = {'File Name':'Unit Count'}, inplace = True)
file_count_df
```

```python
plt.bar(x = file_count_df['Unit'] , height=file_count_df['Unit Count'],
        color=['black', 'red', 'green', 'blue', 'cyan'], edgecolor="black", width=0.5)
plt.xlabel("Units")
plt.ylabel("Unit Count")
plt.show()
```

```python
fig, ax = plt.subplots()
# plt.pie(file_count_df['Unit Count'], labels = file_count_df['Unit'] )
ax.pie(file_count_df['Unit Count'], labels = file_count_df['Unit'], autopct='%1.1f%%')
plt.show()
```

**Number of files in each folder with matplotlib and seaborn**


```python
# visualizing the number of files in each folder with seaborn
sns.countplot(x = 'Folder Name', data = folder_df)
```

```python
# visualizing the number of files in each folder with matplotlib
file_count_df = folder_df.groupby(by=['Folder Name'], as_index=False)['File Name'].count()
file_count_df.rename(columns = {'File Name':'File Count'}, inplace = True)
file_count_df
```

**Folder size comparison with matplotlib and seaborn**

```python
# visualizing the number of files in each folder with matplotlib
folder_size_count_df = folder_df.groupby(by=['Folder Name', 'Folder Size'], as_index=False).count()
folder_size_count_df
```

```python
# visualization with matplotlib
plt.bar(x = folder_size_count_df['Folder Name'] , height=folder_size_count_df['Folder Size'],
        color=['green', 'black', 'yellow', 'red', 'cyan', 'blue'], edgecolor="black", width=0.5)
plt.xlabel("Folders")
plt.ylabel("Size Of Each Folder")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.pie(folder_size_count_df['Folder Size'], labels = folder_size_count_df["Folder Name"], autopct='%1.1f%%')
# plt.pie(folder_size_count_df['Folder Size'], labels = folder_size_count_df["Folder Name"] )
plt.show()
```

```python
# visualization with seaborn
sns.barplot(data = folder_size_count_df, x = "Folder Name", y = "Folder Size")
```

```python
from matplotlib.markers import MarkerStyle
# Line chart folder size visualization
plt.plot(folder_size_count_df['Folder Size'], linewidth = '5', mfc = 'black', marker = 'o', color = 'orange')
plt.title("Folders Size Comparison")
plt.xlabel("Folders")
plt.ylabel("Size of Each Folder")
plt.show()
```
