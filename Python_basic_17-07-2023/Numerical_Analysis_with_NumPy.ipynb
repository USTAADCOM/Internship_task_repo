```python
"""
 Numerical Analysis with Numpy using the BostonHousing.csv data set
"""
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
housing_data = pd.read_csv("sample_data/BostonHousing.csv")
housing_data.head()
```

```python
housing_data.describe()
```

**Creating Numpy array from dataset**

```python
numpy_array = housing_data.to_numpy()
numpy_array.dtype
print(numpy_array)
```

**Calculate the mean, median, and standard deviation of the dataset**

```python
mean = np.mean(numpy_array)
median = np.median(numpy_array)
standard_deviation = np.std(numpy_array)
print("Mean")
print(mean)
print("Median")
print(median)
print("standard_deviation")
print(standard_deviation)
```

**Normalize the data using the z-score normalization technique**
Z-score normalization: This technique scales the values of a feature to have a mean of 0 and a standard deviation of 1. This is done by subtracting the mean of the feature from each value, and then dividing by the standard deviation

```python
plt.figure(figsize=(16,5))
plt.subplot(1,4,1)
sns.histplot(housing_data['age'], kde = True)

plt.subplot(1,4,2)
sns.histplot(housing_data['tax'], kde = True)

plt.subplot(1,4,3)
sns.histplot(housing_data['medv'], kde = True)

plt.subplot(1,4,4)
sns.histplot(housing_data['b'], kde = True)
plt.show()
```

```python
# calculate Z score is calculated on Normally distributed data column
z_score = stats.zscore(housing_data['medv'])
print(z_score)

# calculate Z score using scipy module of numpy array
z_score = stats.zscore(numpy_array)
print(z_score)
```

**Identify outliers in the dataset using the IQR (Interquartile Range) method**
We can use the IQR method of identifying outliers to set up a fence outside of Q1 and Q3. Any values that fall outside of this fence are considered outliers. To build this fence we take 1.5 times the IQR and then subtract this value from Q1 and add this value to Q3. This gives us the minimum and maximum fence posts that we compare each observation to. Any observations that are more than 1.5 IQR below Q1 or more than 1.5 IQR above Q3 are considered outliers

```python
# Outliers are calculated on the basis of data column which are right or left skewed as
# shown above out column b is right skewed
```

```python
housing_data['b'].describe()
```

```python
# It define the following
# 25th percentile
# 50th percentile
# 75th percentile
# Standard Deviation
# Max value
# Min Value
sns.boxplot(data = housing_data["b"])
```

**Finding** the IQR
So from the above steps, we have got the column on which we will apply the method which is best suited for plots that are not normally distributed or the ones whose plot doesnt have the bell curve structure. Then after looking at the boxplot for the same column, we figured out that there are outliers that need to be removed, and for that, here is the start of the section where we will start by finding the IQR,

```python
#  IQR = (75th quartile/percentile  25th quartile/percentile)
percentile25 = housing_data['b'].quantile(0.25)
percentile75 = housing_data['b'].quantile(0.75)
print("75th quartile: ",percentile75)
print("25th quartile: ",percentile25)
```

```python
iqr = percentile75 - percentile25
print ("IQR: ",iqr)
```

```python
print("Lower limit: ")
lower_limit = percentile25 - (1.5 * iqr)
print(lower_limit)
print("Upper limit: ")
upper_limit = percentile75 + (1.5 * iqr)
print(upper_limit)
```

**Finding Outliers**

```python
# Now we will use filter conditions on top of the upper and lower limits so that we will get those rows/tuples which are considered to be bad data points.
housing_data[housing_data['b'] < lower_limit] # Beacuse it lower bound Outliers
```

```python
housing_data[housing_data['b'] < lower_limit].count()
```

**Trimming Data**
Trimming removes all the bad data from the dataset just to ensure

```python
new_data = housing_data[housing_data['b'] > lower_limit]
new_data.shape
```

```python
new_data
```

Inference: Firstly, we returned the outliers rows; then, with the help of the count function, we learned that the total number of rows was 15.

```python
plt.figure(figsize=(16,8))
plt.subplot(2,2,1)
sns.histplot(housing_data['b'], kde=True)

plt.subplot(2,2,2)
sns.boxplot(data = housing_data['b'])

plt.subplot(2,2,3)
sns.histplot(new_data['b'], kde = True)

plt.subplot(2,2,4)
sns.boxplot(data = new_data['b'])

plt.show()
```
