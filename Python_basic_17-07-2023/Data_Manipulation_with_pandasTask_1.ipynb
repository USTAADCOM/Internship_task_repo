**Task 1: Data Manipulation with pandas**
Description: You are given a dataset containing information about sales transactions. Your task is to use pandas to clean and manipulate the data to answer specific questions. Perform the following tasks:

Load the dataset into a pandas DataFrame.
Remove any duplicate rows or missing values.
Extract a subset of columns that are relevant to the analysis.
Calculate the total sales amount for each category.
Identify the top-selling product and its sales quantity.
Generate a summary report with descriptive statistics for numerical columns.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno
```

**Load the dataset into a pandas DataFrame**

```python
sales_df = pd.read_csv("sample_data/sales_data_sample.csv", sep=",", encoding='Latin-1')
sales_df
```

```python
sales_df.describe()
```

**Remove any duplicate rows or missing values**

```python
unique_df = sales_df.drop_duplicates(subset=None, keep="first", inplace=False)
unique_df
```

```python
# Checking for missing values
msno.matrix(unique_df)
```

```python
processedData = unique_df.dropna(axis = "columns", how = "any");
processedData
```

```python
# Checking number of missing values at each column
unique_df.isnull().sum().sort_values(ascending=False)
```

```python
# Checking for after missing values
msno.matrix(processedData)
```

**Extract a subset of columns that are relevant to the analysis**


```python
# # Renaming the field name optionla task
# processedData.rename(index=str, columns={'ORDERNUMBER': 't_id',
#                               'product id': 'product_id',
#                               'product description': 'description',
#                               'quantity sold': 'quantity',
#                               'transaction timestamp': 'timestamp',
#                               'unit price': 'unit_price',
#                               'customer id': 'cust_id',
#                               'transaction country': 'country'}, inplace=True)
```

```python
# Checking for Neagtive values in fileds which never expected negative like quantity
processedData.describe()
```

```python
processedData.drop(['QTR_ID','MONTH_ID','YEAR_ID','PHONE', 'CONTACTLASTNAME', 'CONTACTFIRSTNAME', 'ADDRESSLINE1'], axis = 1, inplace = True)

```

```python
processedData.head()
```

```python
processedData.drop(['DEALSIZE','CITY'], axis = 1, inplace = True)
```

```python
processedData
```

**Calculate the total sales amount for each category**

```python
# Categories in PRODUCTLINE Column
print (processedData['PRODUCTLINE'].unique())
```

```python
# sum_df = processedData.groupby(["PRODUCTLINE"]).sum()
money_spent = processedData.groupby(by=['PRODUCTLINE'], as_index=False)['PRICEEACH'].sum()
money_spent
```

**Identify the top-selling product and its sales quantity**

```python
money_spent.sort_values(by='PRICEEACH', ascending=False).head(1)
```

**Generate a summary report with descriptive statistics for numerical columns**

```python
plt.subplots(figsize=(15,6))
plt.plot(money_spent.PRODUCTLINE , money_spent.PRICEEACH)
plt.xlabel("Products")
plt.ylabel("Total Sales")
plt.title("Sales Of All Products by categorey")
plt.show()
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(13, 11.5)
ax = sns.barplot(x = 'PRODUCTLINE', y ='PRICEEACH', data = money_spent, estimator=max, ax=ax)
ax.set_xticklabels(ax.get_xticklabels(), rotation=47, ha="right")
fig.savefig("Summary.png")
plt.show()
```

**Export the cleaned and processed data to a new CSV file**


```python
processedData.to_csv('processedData.csv')
```
